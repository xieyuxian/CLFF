```python
import numpy as np 
import matplotlib.pyplot as plt 
#手动定义某类标签样本数
num = 10

#生成数据样本
np.random.seed(0) 
x1 = np.random.randn(num,2)-[2,2]    #np.random.randn(num,2)：生成num行2列的服从正态分布的矩阵,-[2,2]表示第1列和第2列所有元素减去2
x2 = np.random.randn(num,2)+[2,2]
x = np.r_[x1,x2]    #np.r_[a,b]：按列连接两个矩阵，即两个矩阵上下拼在一起；记忆：r：row，_：扩展，r_：生成的矩阵行扩展了.通过数组的方式保存这几个随机点
#x = np.r_[np.random.randn(num,2)-[2,2],np.random.randn(num,2)+[2,2]] 
y = [-1]*num+[1]*num #生成标签
```

```python
#画样本散点图
plt.figure()
for i in range(len(x)):
    if y[i] ==1:
        plt.plot(x[i][0],x[i][1],'ro')
    else:
        plt.plot(x[i][0],x[i][1],'bo')
```

```python
#初始化参数
w = np.array([-1,1])
b = 0
delta = 0.01
```

```python
#重绘散点图
plt.figure()
for i in range(len(x)):
    if y[i] ==1:
        plt.plot(x[i][0],x[i][1],'ro')
    else:
        plt.plot(x[i][0],x[i][1],'bo')


#开始训练
for i in range(100):
    #初始化梯度值
    grad_x = np.array([0,0])
    grad_b = 0
    
    for j in range(len(x)):
        #寻找错分样本
        if y[j] != np.sign(np.dot(w,x[j])+b):
            #更新反向梯度值
            grad_x = grad_x+y[j]*x[j]
            grad_b = grad_b+y[j]
            
    #利用梯度下降更新参数
    w = w+delta*grad_x
    b = b+delta*grad_b
    
    #每训练10次更新一次分界线
    if i%10 ==0 : 
        #为了画分界线，必须生成两个点，因为两点确定一条直线
        xx = [-4,4]
        yy = [0,0]
        for i in range(len(xx)):
            yy[i] = -(w[0]*xx[i]+b)/w[1]
            pp = plt.plot(xx,yy)

    
    
        
```

观察可以发现，分界线一步一步地得到优化，最后能达到较好的分类结果
